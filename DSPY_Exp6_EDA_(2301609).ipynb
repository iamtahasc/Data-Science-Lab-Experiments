3.In  Air Quality data set which graph is used to perform multivariate analysis?

**Answer:**

In the Air Quality dataset, the most commonly used graph for multivariate analysis is:

**1. Pair Plot (Scatter Plot Matrix)**

* A pair plot shows relationships between multiple variables at once.

* It creates scatter plots for every pair of numerical features (like PM2.5, PM10, NO2, SO2, CO, etc.).

* Diagonal elements usually show distribution (histogram/KDE) of each variable.

**2. Why it is used:**

* Helps identify correlation between pollutants

* Detects patterns and trends among multiple variables

* Useful for spotting outliers and clusters

**3. Other Multivariate Graphs (sometimes used):**

* Heatmap (Correlation Matrix): Shows strength of relationship between variables

* 3D Scatter Plot: Used when analyzing 3 variables together

4.In Wine quality data set can you do non graphical EDA ? Explain your answer.

**Answer:**

**Yes**, non-graphical EDA can be performed on the Wine Quality dataset.

**Explanation:**

Non-graphical EDA involves analyzing data using statistical and numerical methods without using visualizations. In the Wine Quality dataset, this is effectively done as follows:

**1. Summary Statistics**

* Calculate mean, median, standard deviation, min, max for all features like acidity, pH, alcohol, etc.

* Helps understand overall data distribution.

**2. Measure of Central Tendency**

* Mean and median values of each feature

* Shows typical values of chemical properties in wine.

**3. Measure of Dispersion**

* Variance and standard deviation

* Helps identify how much the values vary.

**4. Correlation Analysis**

* Compute correlation matrix between features and quality

* Helps identify important factors affecting wine quality (e.g., alcohol often positively correlated).

**5. Checking Missing Values**

* Use functions like .isnull().sum()

* Confirms dataset is clean (usually no missing values).

**6. Frequency Distribution of Target Variable**

* Count of each quality score (ordinal values from 0–10)

* Helps understand class imbalance.

**7. Group-wise Analysis**

* Group data based on quality and calculate mean values

* Helps compare chemical properties across different quality levels.
