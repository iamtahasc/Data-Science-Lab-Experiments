# Netflix Data Analysis - Exploratory Data Analysis (EDA)

## Objective
The objective of this notebook is to perform exploratory data analysis using graphical and non-graphical methods.

## Description
EDA helps in understanding patterns, trends, and relationships in the dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cleaned_netflix.csv")
df.head()
```

## Dataset Loaded

The cleaned dataset is loaded successfully and ready for analysis.

```python
df.info()
```

## Non-Graphical Analysis

We perform statistical and numerical analysis without using graphs.

```python
df['type'].value_counts()
```

### Analysis 1: Content Type Distribution

Objective: To check number of Movies and TV Shows  
Observation: Movies are more than TV Shows  
Conclusion: Netflix has more movie content

```python
df['year_added'].value_counts()
```

### Analysis 2: Year-wise Content Addition

Objective: To analyze growth over years  
Observation: Content increased rapidly after 2015  
Conclusion: Netflix expanded globally after 2015

## Graphical Analysis

We use visualizations to understand data patterns.

```python
sns.histplot(df['release_year'], bins=30)
plt.show()
```

### Analysis 3: Release Year Distribution (Univariate)

Objective: To analyze distribution of release years  
Observation: Most content is released after 2000  
Conclusion: Netflix mainly hosts modern content

```python
sns.boxplot(x='type', y='duration_num', data=df)
plt.show()
```

### Analysis 4: Duration vs Type (Multivariate)

Objective: Compare duration of Movies vs TV Shows  
Observation: Movies have fixed duration, TV shows vary  
Conclusion: TV shows have more variation in duration

```python
top = df['country'].value_counts().head(10)

sns.barplot(x=top.values, y=top.index)
plt.show()
```

### Analysis 5: Top Countries

Objective: Identify countries producing most content  
Observation: USA has highest content  
Conclusion: USA dominates Netflix production

## Conclusion

Through EDA, we observed that Netflix mainly focuses on movies, has rapidly expanded after 2015, and contains mostly modern content. USA is the leading contributor of content.
