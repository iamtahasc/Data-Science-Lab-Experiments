**1.Explain input and output characteristics for f_oneway function from stats library.**

**Answer:**



The `f_oneway()` function from the `scipy.stats` library is used to perform a one-way ANOVA test to check whether the means of two or more independent groups are equal.

**Input Characteristics:**
The function takes two or more independent samples as input. Each sample should contain numerical data and can be in the form of a list, tuple, or array. The samples may have different sizes. The data should satisfy certain assumptions such as independence of observations, normal distribution of data, and equal variance across groups.

**Output Characteristics:**
The function returns two values: the F-statistic and the p-value. The F-statistic represents the ratio of variance between the groups to the variance within the groups. A larger F value indicates greater variation between group means. The p-value indicates the probability of obtaining such results if the null hypothesis is true. If the p-value is less than or equal to the chosen significance level (e.g., 0.05), the null hypothesis is rejected; otherwise, it is not rejected.





**2.Explain results of `sm.stats.anova_lm` function**

**Answer:**

The `sm.stats.anova_lm()` function from the `statsmodels` library is used to perform Analysis of Variance (ANOVA) on a fitted linear model. It provides a table that helps in understanding the significance of different factors affecting the dependent variable.

**Results (Output Explanation):**
The function returns an ANOVA table with the following components:

* **df (Degrees of Freedom):**
  Represents the number of independent values associated with each source of variation.

* **sum_sq (Sum of Squares):**
  Measures the total variation attributed to each factor and error. It indicates how much variation is explained by each variable.

* **mean_sq (Mean Square):**
  Calculated as sum of squares divided by corresponding degrees of freedom. It represents average variation.

* **F (F-statistic):**
  Ratio of mean square of a factor to the mean square of error. A higher F value suggests that the factor has a significant effect on the dependent variable.

* **PR(>F) (p-value):**
  Indicates the probability of observing the F-statistic if the null hypothesis is true.

  * If p-value ≤ significance level (e.g., 0.05), the factor is statistically significant.
  * If p-value > significance level, the factor is not statistically significant.

Thus, the `anova_lm()` function helps in determining which independent variables significantly affect the dependent variable by analyzing variance and corresponding statistical measures.



**3.What role ANOVA plays in Data Analytics**

**Answer:**

ANOVA (Analysis of Variance) plays an important role in data analytics by helping analysts compare the means of multiple groups and determine whether the differences among them are statistically significant.

In data analytics, ANOVA is used to identify whether different factors or categories have a meaningful impact on a particular outcome. Instead of performing multiple t-tests, ANOVA provides a single method to test differences across more than two groups, reducing the chances of error.

It helps in:

* **Decision Making:** By determining whether changes in input variables lead to significant changes in output, ANOVA supports data-driven decisions.
* **Feature Importance:** It helps identify which independent variables (factors) significantly affect the dependent variable.
* **Model Validation:** ANOVA is used in regression analysis to evaluate the overall significance of models.
* **Experiment Analysis:** It is widely used in A/B testing, business experiments, and scientific studies to compare multiple conditions or treatments.

Thus, ANOVA is a key statistical tool in data analytics for understanding variation in data, testing hypotheses, and extracting meaningful insights from datasets.




**4.When to use One-Way ANOVA? Explain with real-world case**

**Answer:**

One-Way ANOVA is used when we want to compare the means of **three or more independent groups** based on a single factor (independent variable) to determine whether at least one group mean is significantly different from the others.

It is applied when:

* There is **one independent variable** (factor) with multiple categories.
* There is **one dependent variable** which is continuous (numerical).
* The groups are **independent of each other**.
* The data is approximately **normally distributed** and variances are equal.

**Real-World Case:**
Consider a company that wants to evaluate the effectiveness of **three different training programs** (Program A, Program B, and Program C) on employee performance. After training, the performance scores of employees in each group are recorded.

Here:

* Independent variable: Type of training program (A, B, C)
* Dependent variable: Employee performance scores

One-Way ANOVA is used to test whether the average performance scores differ significantly among the three training programs. If the test shows a significant result, it means that at least one training program is more effective than the others.

Thus, One-Way ANOVA is useful in real-world situations where we need to compare multiple groups based on a single factor to make informed decisions.




**5.When to use Two-Way ANOVA? Explain with real-world case**

**Answer:**

Two-Way ANOVA is used when we want to study the effect of **two independent variables (factors)** on a single continuous dependent variable, and also to check whether there is any **interaction effect** between the two factors.

It is applied when:

* There are **two independent variables**.
* There is **one dependent variable** (numerical).
* The groups are **independent**.
* Data follows **normal distribution** and variances are approximately equal.
* We want to analyze both **individual effects** and **combined (interaction) effect** of factors.

**Real-World Case:**
Suppose a company wants to study how **advertising method** (TV, Social Media) and **region** (Urban, Rural) affect product sales.

Here:

* Independent variable 1: Advertising method
* Independent variable 2: Region
* Dependent variable: Sales amount

Two-Way ANOVA is used to determine:

* Whether the type of advertising affects sales.
* Whether the region affects sales.
* Whether there is an **interaction effect** (e.g., social media ads may work better in urban areas than rural areas).

If the results are significant, the company can optimize its marketing strategy based on both factors.

Thus, Two-Way ANOVA helps in understanding the combined influence of two factors on an outcome, making it highly useful in data analytics and decision-making.
